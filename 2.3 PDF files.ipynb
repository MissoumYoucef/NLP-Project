```python
f = open('pdffile.pdf','rb')

# List of every page's text.
# The index will correspond to the page number.
pdf_text = [0]  # zero is a placehoder to make page 1 = index 1

pdf_reader = PyPDF2.PdfReader(f)

for p in range(len(pdf_reader.pages)):
    
    page = pdf_reader.pages[p]
    
    pdf_text.append(page.extract_text())

f.close()

print(pdf_text)
```

```python
f = open('pdffile.pdf','rb')
pdf_reader = PyPDF2.PdfReader(f)

first_page = pdf_reader.pages[0]
first_page.extract_text()
```

```python
pdf_writer = PyPDF2.PdfWriter()
```

```python
pdf_writer.add_page(first_page)
```

```python
pdf_output = open("new file.pdf","wb")
```

```python
pdf_writer.write(pdf_output)
```

```python
pdf_output.close()
```

```python
from PyPDF2 import PdfMerger, PdfReader,  PdfWriter

filename1 = 'pdffile.pdf'
filename2 = 'new file.pdf'

merger = PdfMerger()

merger.append(PdfReader(open(filename1, 'rb')))
merger.append(PdfReader(open(filename2, 'rb')))

merger.write("merged file 1.pdf")
```

```python
file1 = PdfReader(open(filename1, "rb"))
file2 = PdfReader(open(filename2, "rb"))

output = PdfWriter()

output.add_page(file1.pages[5])
output.add_page(file1.pages[3])


outputStream = open("merged file 2.pdf", "wb")
output.write(outputStream)
outputStream.close()
```
